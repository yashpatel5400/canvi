```python
import pickle
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
outputdir = "output/"
simulations = [2**n for n in range(10, 18)]
```

```python
coverages = []
mis = []

for sim in simulations:
    resultdir = os.path.join(outputdir, str(sim))
    coverages.append(np.load(os.path.join(resultdir, "coverage.npy")))
    mis.append(np.load(os.path.join(resultdir, "mutual_information.npy")))
```

```python
plt.plot(simulations, coverages)
plt.xlabel("Simulation budget")
plt.ylabel("Coverage")
plt.xscale("log")
plt.show()
```

```python
plt.plot(simulations, mis)
plt.xlabel("Simulation budget")
plt.ylabel("Mutual information")
plt.xscale("log")
plt.show()
```
